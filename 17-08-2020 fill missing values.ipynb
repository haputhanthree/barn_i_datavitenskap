```python
# Importing packages

import pandas as pd
```

```python
# Reading data from csv file

dataPath = "winequalityN.csv"
```

```python
df = pd.read_csv(dataPath)
```

```python
# Viewing the data

df.head()
```

```python
#rename columns space to underscore
df = df.rename(columns={'fixed acidity': 'fixed_acidity', 'volatile acidity': 'volatile_acidity' 
                        , 'citric acid' : 'citric_acid','residual sugar' : 'residual sugar'
                       ,'free sulfur dioxide': 'free_sulfur_dioxide' ,'total sulfur dioxide':'total_sulfur_dioxide'})
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isna().sum()
```

## Replace missing values with mean

```python
# #fill missing value with mean single column

# df.fillna({"fixed acidity":df['fixed acidity'].mean()}, inplace=True)
```

```python
df[['fixed_acidity', 'volatile_acidity']] = df[['fixed_acidity', 
                                                'volatile_acidity']].fillna(df[['fixed_acidity', 
                                                                                'volatile_acidity']].mean())
```

```python
df.isna().sum()
```

## Remove Missing value rows

```python
# The dataset has null values, this will remove the all missing value rors

df.dropna(inplace=True)
```

```python
df.isna().sum()

# The dataset is clean and free from null values
```
