```python
# df[['fixed_acidity', 'volatile_acidity']] = df[['fixed_acidity', 
#                                                 'volatile_acidity']].fillna(df.groupby('type')['fixed_acidity', 
#                                                                                 'volatile_acidity'].mean())
```

```python
df[['fixed_acidity', 'volatile_acidity','citric_acid','residual_sugar','free_sulfur_dioxide','total_sulfur_dioxide']] = df.groupby(['type'])[['fixed_acidity', 'volatile_acidity','citric_acid','residual_sugar','free_sulfur_dioxide','total_sulfur_dioxide']]\
    .transform(lambda x: x.fillna(x.mean()).round(1))
```

```python
df[['fixed_acidity', 'volatile_acidity'
    ,'citric_acid','residual_sugar','chlorides','pH'
    ,'sulphates']] = df.groupby(['type'])[['fixed_acidity', 'volatile_acidity'
                                                      ,'citric_acid','residual_sugar','chlorides','pH'
                                                      ,'sulphates']].transform(lambda x: x.fillna(x.mean()).round(1))
```
